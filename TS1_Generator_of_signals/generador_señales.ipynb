En este primer trabajo comenzaremos por diseñar un generador de señales que utilizaremos en las primeras simulaciones que hagamos. La primer tarea consistirá en programar una función que genere señales senoidales y que permita parametrizar:

* La amplitud máxima de la senoidal (volts)
* Su valor medio (volts)
* La frecuencia (Hz)
* La fase (radianes)
* La cantidad de muestras digitalizada por el ADC (# muestras)
* La frecuencia de muestreo del ADC.

Es decir que la función que ustedes armen debería admitir ser llamada de la siguiente manera:

  `tt, xx = mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = N, fs = fs)`

Recuerden que tanto **xx** como **tt** deben ser vectores de Nx1. Puede resultarte útil el módulo de visualización **matplotlib.pyplot** donde encontrarán todas las funciones de visualización estilo Matlab. Para usarlo:

 `import matplotlib.pyplot as plt`
 
 `plt.plot(tt, xx)`
***

### Solucion propuesta:
***
##### Modulos utilizados:

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Funcion requerida:

```python
def mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = 1, fs = 1):
    ts = 1/fs
    
    tt = np.linspace(0, (nn-1)*ts, nn).flatten()
    
    x = np.array([], dtype=np.float).reshape(nn,0)
    
    aux = vmax * np.sin(2*np.pi*ff*tt + ph) + dc
    
    x = np.hstack([x, aux.reshape(nn,1)] )
    
    plt.figure(1, figsize=(10, 6), dpi=100)
    
    line_hdls = plt.plot(tt, x, label=f"Vmax: {vmax}v\nVmed: {dc}v\nFrec.: {ff}Hz\nFase: {ph}rad")
    
    plt.grid(color='gray', linestyle='dotted', linewidth=0.8)
    
    plt.title('Generador de Señales' )
    
    plt.xlabel('tiempo [segundos]')
    
    plt.ylabel('Amplitud [V]')
    
    plt.legend(loc='upper right', shadow=True, fontsize='small')
    
    plt.show()
```

##### Ejemplo de utilizacion:

```python
mi_funcion_sen(4, 4, 10, round(np.pi, 3), 1000, 1000)
```
